```python
# Time Series - Past Data
# Forcasting - predicting future 

# A time series is a sequence of numerical data points in successor order

# Component :
# describe the historical patterns in the data
# prepare forecast by projecting the historical pattern furture

regression = House price / total area / room size / poln details / crime rate

# Additive model
Time     Revenue
Q1  = 120
Q2  = 150
Q3  = 180
Q4  = 210
Q5  = 240
Q6  = ???

#multiplicative model
Time     Revenue
Q1  = 100
Q2  = 225
Q3  = 400
Q4  = 625
Q5  = 900
Q6  = ???


Time     Revenue
Q1  = 100
Q2  = 105
Q3  = 400
Q4  = 900
Q5  = 1685
Q6  = ???

# There are 4 components we have in Time Series
# Trends - data is increasing or decreasing during a long period of time
# Seasonality - this occur at a specific regular intervals less than a year, 
        #such as weekly, monthly, or quarterly
# Cyclic components - A cyclic pattern exists when data rises or falls 
        # that are not fixex at a certain period - uncertainity
    
# Unexpected / Noisy / Random / Irregular components - erratic / uncontrollable / unforeseen
    # earthquakes, wars, flood, disasters, catastrophe etc.........
# Noisy data - there is no correlation - null "0"
          

# Type of Time Series - Two types
# Additive time series
Yt = Tt + St + Ct + It

where Yt = predict the value (future value)

# Multiplicative time series
Yt = Tt * St * Ct * It

where Yt = predict the value (future value)


# Time Series Model - 
# Different Methods of Time Series forecasting as follow - 
**1) AR - AutoRegressive Model
**2) MA - Moving Average
**3) ARMA - Autoregressive and Moving Average
*****4) ARIMA - Autoregressive Integrated Moving Average - Very important model 
**5) SARIMA - Seasonal Autoregressive Integrated Moving Average
**6) ARIMAX - Autoregressive Integrated Moving Average with external regressor
7) SARIMAX - Seasonal Autoregressive Integrated Moving Average with external regressor
8) VAR - Vector autoregressive
9) VARMA - Vector Autoregressive and Moving Average
10) VARMAX - Vector Autoregressive and Moving Average with external regressor
**11) SES - Simple Exponetial Smoothing - "Winter's Model" - "1960 model"
**12) Double Exponential Smoothing - Holts model - 1957 model
**13) Holt and winter's exponential smoothing (HWES) / 
    #Simple and double exponential smoothing / muliptiple exponential smoothing
14) and many more model but not important


# Measurement of Time Series Forecasting - Accuracy / Error

1) MAE - Mean Absolute Error
# Yt = Actual value
# Ft = Future value
# MAE is less than model accuracy is very good.


2) MAPE - Mean Absolute Percent Error


# MAPE is less then model accuracy is good

3) MSE - Mean Square Error - 1/n * summation of(actual-predicted)^2

4) RMSE - Root Mean Square Error

5) AIC - Akike Information Criterion

AIC = -2LL + 2K 
where, LL = Log Likelihood and K = no. of parameter

# time series concept comes here,when your parameter "K"
LL = log likelihood,K=parameter
K=no. of parameter ; parameter= pdq 
where p=partial auto correlation (AR MODEL)
d = integrated value (I model)
q = auto correlation (MA Model)

ARIMA - This is the main algorithm to find time series forecasting 


# two types of parameter
PDQ and pdq 
PDQ = Trend
pdq = seasonality



6) BIC - Bayesian Information Criteria

BIC = -2LL + K*Ln(n)
where, LL = Log likelihood
K = Parameters = PDQ (Trend) and pdq (seasonality)
Ln = Natural Log
n = no. of observation


# Time Series forecastiong can perform only when your data is stationary
# Stationary ??
# Hypothesis Testing - 
# The intuition behind the test that is determines how strongly a time series is defined by a trend
# If your p-value is alternative, data is stationary

# H0 : Null Hypothesis  - Not Stationary
# Ha : Alternative Hypotheis - P-value <=0.05, reject the null hypothesis, and perform time series forcasting
# p-value >0.05, Fail to reject null hypothesis, it's non-statationary data. 
    #hence, you can not perforem time series forecasting
    

######################################
#######Maths formula
1) Winter's model - 1960 - singal exponetial smoothing
2) Holt Model - double exponetial smoothing



```
